```python
a=[1,2,3,4,5]
b=[2,4,5,6,7]
c=[3,4,5,6,7]
plt.scatter(a,b)
```

```python
plt.plot(a,b,marker='o',label='cube',color='red')
plt.plot(b,c,marker='*',label='square')
plt.legend() # for label
```

```python
plt.plot(a,b,marker='o',label='cube',color='red',alpha=0.1) # alpha is used to adjust the sharpness of line
plt.plot(b,c,marker='*',label='square')
plt.axis('off') # for removing axis
plt.legend()
plt.title('square vs title') # for title
plt.show()
```

```python
plt.plot(a,b,marker='o',label='cube',color='red',alpha=0.1) 
plt.plot(b,c,marker='*',label='square',linestyle='dotted') # linestyle is usd to form diff kinds of line
plt.axis('off')
plt.legend()
plt.title('square vs title') # for title
plt.show()
```

```python
a=["maths","chem","eng","physics"]
b=[40,56,45,23]
plt.pie(b,labels=a,autopct='%0.3f%%',shadow="true")
```
